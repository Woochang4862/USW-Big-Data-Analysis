```python
import os
import psutil


# 파일 읽기 및 단어 빈도 계산
IF = open("../data/ratings_train.txt", "r", encoding='UTF8')
dic_word2freq = {}
for line in IF:
    s = line.strip().split("\t")
    if len(s) < 2:  # 빈 줄 방지
        continue
    lst_word = s[1].split(" ")
    for word in lst_word:
        if word not in dic_word2freq:
            dic_word2freq[word] = 0
        dic_word2freq[word] += 1
IF.close()


# 단어 빈도 상위 10개 출력
lst_word_freq = sorted(dic_word2freq.items(), key=lambda x: x[1], reverse=True)
print("상위 10개 단어:", lst_word_freq[:10])


# 전체 시스템 메모리 상태 출력
mem = psutil.virtual_memory()
print(f"\n[시스템 메모리 정보]")
print(f"전체 용량: {mem.total / (1024 ** 3):.2f} GB")
print(f"사용 중: {mem.used / (1024 ** 3):.2f} GB")
print(f"사용률: {mem.percent}%")


# 현재 파이썬 프로세스의 메모리 사용량 출력
process = psutil.Process(os.getpid())
mem_info = process.memory_info()
print(f"\n[파이썬 프로세스 메모리 정보]")
print(f"RSS (실제 사용 중인 물리 메모리): {mem_info.rss / (1024 ** 2):.2f} MB")
print(f"VMS (가상 메모리): {mem_info.vms / (1024 ** 2):.2f} MB")
```
