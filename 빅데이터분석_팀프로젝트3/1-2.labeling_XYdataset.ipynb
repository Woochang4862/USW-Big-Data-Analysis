```python
import pandas as pd
import numpy as np
df = pd.read_excel('hn2223_all.filtered.xlsx', index_col=0)
```

```python
df
```

```python
# age 범주화: 19 ~ 64세까지를 '19-64', 65세 이상을 '65+'로 구분
age_group = pd.cut(df['age'], bins=[19, 64, float('inf')], labels=['19-64', '65+'])
pd.crosstab(df['sex'], 
                 age_group, 
                 margins=True,            # 합계 추가
                 margins_name='Total')     # 합계 행/열 이름
```

```python
df = df[~df['HE_D3'].isna()]  # nan(결측치) 제외
print(df.shape)
df['HE_D3_label'] = df['HE_D3'].apply(lambda x: 1 if x > 20 else 0)  # 라벨링 비타민D결핍=0, 충분=1
print(df.shape)
```

```python
# age 범주화: 19 ~ 64세까지를 '19-64', 65세 이상을 '65+'로 구분
age_group = pd.cut(df['age'], bins=[19, 64, float('inf')], labels=['19-64', '65+'])
pd.crosstab(
    [df['sex'], age_group],
    df['HE_D3_label'], 
    margins=True,
    margins_name='Total'
)
```

```python
# 제외할 8개 컬럼
exclude_cols = ['HE_D2', 'HE_D3_3epi', 'HE_D3'] # 혈중 비타민D 변수 HE_D3_label와 직접 관련된 covariates 변수 제거

# X (특정 8개 컬럼 제외한 모든 컬럼)
X = df.drop(columns=exclude_cols + ['HE_D3_label'])

# Y (타겟)
Y = df['HE_D3_label']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

# 데이터 분할 (stratify는 HE_D3_label 분포가 중요할 때 권장)
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)

# TSV 파일로 저장
X_train.to_csv('X_train.csv', index=True)
X_test.to_csv('X_test.csv', index=True)
Y_train.to_csv('Y_train.csv', index=True)
Y_test.to_csv('Y_test.csv', index=True)
```

```python
sample_submission = Y_test
sample_submission[:] = 1
sample_submission.to_csv('sample_submission.csv', index=True)
```
