```python
import pandas as pd
import numpy as np
df22 = pd.read_sas('./hn22_all.sas7bdat')
df23 = pd.read_sas('./hn23_all.sas7bdat')
df_meta = pd.pandas.read_excel('./updated_dataframe.xlsx')
```

```python
# 1. 공통 컬럼(교집합) 추출
common_cols = set(df22.columns) & set(df23.columns)

# 2. df23의 컬럼 순서를 기준으로 공통 컬럼 정렬
ordered_cols = [col for col in df23.columns if col in common_cols]

# 2. 세로 방향으로 병합
df = pd.concat([df22[list(ordered_cols)], df23[list(ordered_cols)]], axis=0, ignore_index=True)
df.set_index('ID', inplace=True)
df
```

```python
df_meta
```

```python
# df_mata의 'option description'에 따라 비해당 값은 -1로 모름/무응답은 NaN으로 치환
for idx, row in df_meta.iterrows():
    # 변수를 선택
    variable = row['variable']
    if variable in df.columns and pd.notna(row['option description']):
        # 변수의 선택지를 파싱 e.g.) 1:예 | 2: 아니오 | 8:비해당 | 9 : 모름/무응답
        # 변수에 따라 비해당과 모름/무응답 값은 8,9 또는 88,99 또는 888,999 등이 될 수 있음
        options = dict(item.split(':') for item in row['option description'].split('|'))
        for key, value in options.items():
            # df의 비해당 값을 -1로 치환
            if ('8' in key or '9' in key) and '비해당' in value:
                df[variable] = df[variable].replace(float(key), -1)
            # df의 모름/무응답 값을 NaN 값으로 치환
            elif ('8' in key or '9' in key) and ('모름' in value or '무응답' in value):
                df[variable] = df[variable].replace(float(key), np.nan)
```

```python
# 분석에 사용하지 않는 컬럼들을 제거
drop_columns = list(df_meta[df_meta['data type'] == 'object']['variable']) # object 타입(문자열 타입) 제거
drop_columns += list(df_meta[df_meta['drop_col'] == 1]['variable']) # 통계 샘플에 대한 가중치 변수 제거
df = df.drop(columns=drop_columns, errors='ignore')
df
```

```python
#'N_DUSUAL' 컬럼에서 b'.'이라는 NaN의 문제를 해결

# 변수들의 type을 확인 --> object 변수를 모두 제거했음에도 한 개의 object 타입 변수가 존재
print(df.dtypes.value_counts(dropna=False))

# 어떤 변수가 object 타입 변수인지 확인 --> N_DUSUAL 변수임을 확인
print(f'\nobject columns : ', df.columns[df.dtypes == 'object'].tolist())

# 어떤 문자 값이 있는지 확인 --> 점 문자 b'.' 때문에 문자열 값으로 인식하는 문제 발생
print(df['N_DUSUAL'].value_counts(dropna=False))

# 'N_DUSUAL' 컬럼에서 b'.' 값을 NaN으로 치환
df['N_DUSUAL'] = df['N_DUSUAL'].replace(b'.', np.nan)

# 'N_DUSUAL' 컬럼의 데이터 타입을 float64로 변경
df['N_DUSUAL'] = df['N_DUSUAL'].astype('float64')

# 'N_DUSUAL' 컬럼의 값 확인 --> 모두가 실수형으로 바뀜
print(df['N_DUSUAL'].value_counts(dropna=False))

# 변수들의 type을 확인 --> 모두 float64 타입이며, 더 이상 object 타입 변수가 없음
print(df.dtypes.value_counts(dropna=False))
```

```python
# 연령 19세 미만인 사람 제거
df = df[df['age'] >= 19]

# age 범주화: 19 ~ 65세까지를 '19-65', 65세 이상을 '65+'로 구분
age_group = pd.cut(df['age'], bins=[19, 65, float('inf')], labels=['19-65', '65+'])
pd.crosstab(df['sex'], 
                 age_group, 
                 margins=True,            # 합계 추가
                 margins_name='Total')     # 합계 행/열 이름
```

```python
# 참여자 별로 NaN 값의 비율에 대한 히스토그램, 변수별 NaN 값의 비율에 대한 히스토그램 확인

import pandas as pd
import matplotlib.pyplot as plt

# 그림과 축을 생성합니다. (2개의 축, 수직 배치)
fig, axs = plt.subplots(3, 1, figsize=(5, 12))  # 두 개의 플롯, figsize를 조정하여 전체 그림 크기 설정

# 첫 번째 플롯: 각 참가자별 NaN 값의 비율
nan_counts_parts = df.isna().sum(axis=1)
nan_percentage_parts = (nan_counts_parts / df.shape[1]) * 100  # 참가자 별 NaN 계산을 위해 df의 컬럼 수로 나눔
axs[0].hist(nan_percentage_parts, bins=100, color='red', alpha=0.7)
axs[0].set_title(f'Distribution of NaN Values Per Participant (N={df.shape[0]})')
axs[0].set_xlabel('Percentage of NaN Values')
axs[0].set_ylabel('Number of Participants')
axs[0].grid(True)

# 두 번째 플롯: 각 변수별 NaN 값의 비율
nan_counts_vars = df.isna().sum(axis=0)
nan_percentage_vars = (nan_counts_vars / len(df)) * 100
axs[1].hist(nan_percentage_vars, bins=100, color='blue', alpha=0.7)
axs[1].set_title(f'Distribution of NaN Values Per Variable (V={df.shape[1]})')
axs[1].set_xlabel('Percentage of NaN Values')
axs[1].set_ylabel('Number of Variables')
axs[1].grid(True)

# 세 번째 플롯: 전체 값의 분포
df_numeric = df.apply(pd.to_numeric, errors='coerce')

all_values = df_numeric.values.flatten()
all_values = all_values[~np.isnan(all_values)]

axs[2].hist(all_values, bins=100, edgecolor='black', range=(-2, 100))
axs[2].set_title('Histogram of All Values in DataFrame')
axs[2].set_xlabel('Value')
axs[2].set_ylabel('Frequency')
axs[2].grid(axis='y', alpha=0.75)

stats_text = (
    f'Min: {np.min(all_values):.2f}\n'
    f'Max: {np.max(all_values):.2f}\n'    
    f'Mean: {np.mean(all_values):.2f}\n'
    f'Median: {np.median(all_values):.2f}\n'
    f'N(<0): {np.sum(all_values < 0):d} ({np.sum(all_values < 0)/df.size*100:.1f}%)\n'
    f'N(nan): {df_numeric.isna().sum().sum()} ({df_numeric.isna().sum().sum()/df.size*100:.1f}%)'
)
axs[2].legend([stats_text], loc='upper right')

# 전체 그림을 표시합니다.
plt.tight_layout()  # 서브플롯 간격을 자동으로 조정
plt.show()
```

```python
# 변수 중에서 NaN 값의 비율이 80% 이상인 열을 제거

# 변수별 NaN 값의 비율을 계산
variable_nan_percentage = df.isna().sum(axis=0) / df.shape[0]

# 변수 중에서 NaN 값의 비율이 80% 이상인 열을 제거
df = df.loc[:, variable_nan_percentage < 0.8]
df
```
