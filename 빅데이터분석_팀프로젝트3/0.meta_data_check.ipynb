```python
import pandas as pd
df = pd.pandas.read_excel('./국건영(지침서)_2023년최신.xlsx')
```

```python
df
```

```python
# 모든 컬럼에 대해서 값에 좌우 공백 제거 strip() 적용
for col in df.columns:
    df[col] = df[col].map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# option description 컬럼에 대해서 값에 좌우 공백 제거 strip() 적용
for index, row in df.iterrows():
    option_desc = row['option description']
    if pd.isna(option_desc):  # NaN 값 체크
        continue  # NaN 값인 경우 다음 행으로 넘어갑니다.
    try:
        # "|"로 구분한 후 각 항목에 대해 strip 적용
        options = [opt.strip() for opt in str(option_desc).split('|') if opt.strip()]
    except AttributeError as e:
        print(f"에러: '{option_desc}' ({type(option_desc)}) 객체는 split 메소드를 지원하지 않습니다. 행 인덱스: {index}")            
        
    updated_options = []
    for opt in options:
        if ':' not in opt:
            if row['data type'] == 'category' :
                print(f"에러: '{opt}' 에서 ':'가 누락되었습니다. 행 인덱스: {index}")
            else:
                print(f"주의: {row['data type']} : '{opt}' 에서 ':'가 누락되었습니다. 행 인덱스: {index}")                
        else:
            parts = opt.split(':')
            if len(parts) != 2:
                if row['data type'] == 'category' :
                    print(f"에러: '{opt}' 에서 ':'가 예상과 다르게 사용되었습니다. 행 인덱스: {index}")
                else:
                    print(f"주의: {row['data type']} : '{opt}' 에서 ':'가 예상과 다르게 사용되었습니다. 행 인덱스: {index}")    
            else:
                key, value = parts[0].strip(), parts[1].strip()
                updated_options.append(f"{key}:{value}")
    
    # 업데이트된 옵션 문자열을 다시 결합
    df.at[index, 'option description'] = '|'.join(updated_options)
```

```python
output_file = 'updated_dataframe.xlsx'
df.to_excel(output_file, index=False)
```
